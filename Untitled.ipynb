```python
from bs4 import BeautifulSoup
import lxml
```

```python
r = requests.get('https://enchantedlearning.com/wordlist/animal.shtml')
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
l = []
```

```python
for item in soup.find_all('div', class_='wordlist-item'):
    try:
        if type(item.find('a').text):
            text = item.find('a').text
            if len(text) < 6:
                if ' ' not in text:
                    l.append(text)
    except AttributeError as e:
        pass
```

```python
final_list = []
for item in l:
    if len(item) < 6:
        if ' ' not in item:
            final_list.append(item)
```
